```python
# Agora vamos trabalhar como cientista de dados, o cara que vai tratar os dados, modificar o banco de dados, identificar alguns problemas e eliminá-los
#Vamos fazer tratamento de dados faltantes, os missings

import pandas as pd
```

```python
#Vamos criar a seleção para passar para o dataframe

# dados['Valor'].isnull() Aqui é criada uma series booleana, mostrando true ou false quando é nulo ou não nulo, respectivamente.

# Dentro de dados, eles retornam apenas os que são nulos, ou seja, o que é TRUE

dados[dados['Valor'].isnull()]
```

```python
# No caso do valor, não queremos valores nulos no nosso banco, ele não serve para nada. A variável do nosso banco é o valor do aluguel, é com ele que veremos o tipo de imóvel é, suas características, o porquê ele tem esse valor, etc

# Nesse caso, quero eliminar eles do banco

# Vamos ver o tamanho do dataframe antes de eliminarmos e depois de eliminar-mos

# Mexemos no dataframe e depois salvamos ele no csv para utilizarmos posteriormente
A = dados.shape[0]
A
```

```python
# O argumento subset é a lista com as variáveis ou variável que queremos apagar
dados.dropna(subset = ['Valor'], inplace = True )
```

```python
B = dados.shape[0]
A - B

# o resultado abaixo é o número de registros eliminados do dataframe
```

## Tratamento de Dados Faltantes (Continuação)

```python
# # Temos condomínio e iptu nulos, mas o condomínio tem mais problema ainda, porque o que acontece:
# # Condomínio nulo pra casa = ok, não quero eliminar eles
# # Assumindo que todos os condomínios têm condomínio, se ele for nulo, eu elimino
# #Vendo quantos valores de condomínio eu tenho nulo:

# dados[dados['Condominio'].isnull()].shape[0]

dados[dados['Condominio'].isnull()].shape[0]
```

```python
# # Vamos fazer uma seleção onde eu tenho imóveis do tipo apartamento, que tem condomínio nulo

# selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
# selecao
# # São justamente os que eu não quero, que quero apagar. 


selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())

selecao
```

```python
# dados [selecao] nos dá dados que são tipos apartamento e tem condomínio nulo, que eu quero eliminar, para inverter, coloco ~ na frente para inverter essa seleção que é uma series booleana.
# invertendo: dados [~selecao], o que é true vira falso, o que é falso, vira true


# A = dados.shape[0]
# dados = dados[selecao]
# B = dados.shape[0]
# A-B

A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
# Continua com:
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Agora dados com Condomínio null e iptu null, vou manter no banco, mas irei atribuir valor a eles


# Forma 1, mas altero de todo o dataframe dados.fillna(0, inplace = True)


# Outra forma de fazer isso seria criar outro dataframe, um dicionário, onde passarei o nome da variável e vou passar o valor que quero que substitua. 
# Isso é legal porque posso passar valores diferentes para variáveis diferentes

dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados[dados['Condominio'].isnull()].shape[0]


# vamos ver a quantidade de nulls que tenho agora, se for 0 = ok, correto
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

```python
dados.info()
```

```python
#todas as informações agora estão não nulas

#Agora vamos salvar isso e subtituir esse arquivo no arquivo aluguel_residencial porque não preciso do dataframe do jeito que ta:


dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
